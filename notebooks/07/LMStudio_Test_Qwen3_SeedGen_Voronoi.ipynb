```python
# run LM Studio with local server enabled
# Batch-run N cases with LM Studio *grammar / schema constrained* output via response_format=json_schema
# - For each case: print INPUT_JSON + OUTPUT_JSON
# - Then display the Voronoi figure directly as cell output (one figure per case)

%pip -q install -U numpy matplotlib requests shapely

import json, random
import requests
import numpy as np
import matplotlib.pyplot as plt

from shapely.geometry import MultiPoint, box
from shapely import voronoi_polygons

LMSTUDIO_URL = "http://localhost:1234/v1/chat/completions"
MODEL_NAME = "local-model"

# -----------------------------
# Input + prompt + call
# -----------------------------
def make_input(aspect_ratio=1.0, k=6, hotspot_count=8, seed=0):
    rng = np.random.default_rng(seed)
    random.seed(seed)
    ws = rng.dirichlet([0.8] * hotspot_count).tolist()
    hotspots = []
    for i in range(hotspot_count):
        hotspots.append({
            "x": random.random(),
            "y": random.random(),
            "weight": float(ws[i]),
            "radius": float(random.uniform(0.02, 0.10)),
        })
    return {"aspect_ratio": float(aspect_ratio), "k": int(k), "hotspots": hotspots}

def build_prompt(inp):
    k = inp["k"]
    return (
        "Place shard seed centers in a normalized 2D world.\n"
        "Coordinates are in [0,1]x[0,1]. The world aspect ratio is width/height.\n\n"
        f"INPUT_JSON:{json.dumps(inp, separators=(',',':'))}\n\n"
        "Return ONLY JSON (no extra text).\n"
        f"Return exactly {k} centers.\n"
        "Centers should be unique (no duplicates)."
    )

def make_response_schema_for_k(k: int):
    return {
        "name": "voronoi_centers",
        "schema": {
            "type": "object",
            "additionalProperties": False,
            "properties": {
                "centers": {
                    "type": "array",
                    "minItems": int(k),
                    "maxItems": int(k),
                    "items": {
                        "type": "object",
                        "additionalProperties": False,
                        "properties": {
                            "x": {"type": "number", "minimum": 0.0, "maximum": 1.0},
                            "y": {"type": "number", "minimum": 0.0, "maximum": 1.0},
                        },
                        "required": ["x", "y"],
                    },
                }
            },
            "required": ["centers"],
        },
    }

def call_model_constrained(prompt, k: int, temperature=0.2, timeout=120):
    schema = make_response_schema_for_k(k)
    payload = {
        "model": MODEL_NAME,
        "messages": [{"role": "user", "content": prompt}],
        "temperature": float(temperature),
        "response_format": {"type": "json_schema", "json_schema": schema},
    }
    r = requests.post(LMSTUDIO_URL, json=payload, timeout=timeout)
    r.raise_for_status()
    return r.json()["choices"][0]["message"]["content"]

def extract_output_json_constrained(text):
    return json.loads(text.strip())

# -----------------------------
# Shapely Voronoi helpers
# -----------------------------
def _dedupe_points(pts: np.ndarray, eps: float = 1e-10) -> np.ndarray:
    # Quantize to grid then unique (prevents pathological duplicates)
    q = np.round(pts / eps) * eps
    _, idx = np.unique(q, axis=0, return_index=True)
    return pts[np.sort(idx)]

def _jitter_points(pts: np.ndarray, scale: float, seed: int) -> np.ndarray:
    rng = np.random.default_rng(seed)
    return pts + rng.normal(scale=scale, size=pts.shape)

def _sanitize_points_for_shapely(
    pts: np.ndarray,
    xmin: float, xmax: float, ymin: float, ymax: float,
    seed: int = 0,
) -> np.ndarray:
    # Shapely/GEOS is more tolerant than Qhull but still benefits from de-dupe.
    pts_u = _dedupe_points(pts, eps=1e-10)

    if pts_u.shape[0] < 2:
        # If the model produced identical centers, jitter to get something drawable.
        pts_u = _jitter_points(pts, scale=1e-6, seed=seed)
        pts_u = _dedupe_points(pts_u, eps=1e-12)

    pts_u[:, 0] = np.clip(pts_u[:, 0], xmin, xmax)
    pts_u[:, 1] = np.clip(pts_u[:, 1], ymin, ymax)
    return pts_u

def _shapely_voronoi_polys(pts_xy: np.ndarray, xmin, xmax, ymin, ymax, tolerance=1e-10):
    """
    Returns polygons clipped/extended to bbox.
    If input is degenerate, this may raise; caller can retry with jitter.
    """
    extent = box(xmin, ymin, xmax, ymax)
    mp = MultiPoint([tuple(p) for p in pts_xy])

    gc = voronoi_polygons(
        mp,
        extend_to=extent,     # ensures we cover the whole rectangle
        tolerance=tolerance,  # snaps nearly-identical inputs
        only_edges=False
    )

    polys = list(gc.geoms) if hasattr(gc, "geoms") else list(gc)
    # Clip each polygon to bbox (extend_to usually handles, but clipping is cheap + safe)
    out = []
    for p in polys:
        if p.is_empty:
            continue
        out.append(p.intersection(extent))
    return out

# -----------------------------
# Plot
# -----------------------------
def plot_voronoi(inp, out_json):
    aspect = float(inp["aspect_ratio"])
    centers = out_json["centers"]
    hotspots = inp["hotspots"]

    # Convert normalized [0,1]x[0,1] to world coords [0,aspect]x[0,1]
    pts = np.array([[c["x"] * aspect, c["y"]] for c in centers], dtype=float)
    xmin, xmax, ymin, ymax = 0.0, aspect, 0.0, 1.0

    pts_u = _sanitize_points_for_shapely(pts, xmin, xmax, ymin, ymax, seed=0)

    polys = None
    # Retry with increasing jitter if GEOS rejects degenerate inputs
    for attempt, jitter in enumerate([0.0, 1e-12, 1e-9, 1e-6, 1e-4]):
        try_pts = pts_u if jitter == 0.0 else _sanitize_points_for_shapely(
            _jitter_points(pts_u, scale=jitter, seed=attempt),
            xmin, xmax, ymin, ymax, seed=attempt
        )
        try:
            polys = _shapely_voronoi_polys(try_pts, xmin, xmax, ymin, ymax, tolerance=1e-10)
            pts_u = try_pts
            break
        except Exception:
            polys = None
            continue

    if polys is None:
        raise RuntimeError("Shapely Voronoi failed; centers too degenerate even after jitter.")

    fig, ax = plt.subplots(figsize=(7.5, 5.5))
    ax.set_aspect("equal", adjustable="box")
    ax.set_xlim(xmin, xmax)
    ax.set_ylim(ymin, ymax)

    for poly in polys:
        if poly.is_empty:
            continue
        # Handle MultiPolygon from intersection
        geoms = list(poly.geoms) if poly.geom_type == "MultiPolygon" else [poly]
        for g in geoms:
            x, y = g.exterior.xy
            ax.fill(x, y, alpha=0.15, linewidth=0)
            ax.plot(x, y, linewidth=1)

    ax.scatter(pts_u[:, 0], pts_u[:, 1], marker="x")

    for h in hotspots:
        hx, hy = h["x"] * aspect, h["y"]
        ax.scatter([hx], [hy], s=2000 * float(h["weight"]), alpha=0.5)
        ax.add_patch(plt.Circle((hx, hy), float(h["radius"]) * aspect, fill=False, linewidth=1))

    ax.set_title(f"Voronoi (Shapely) | k={inp['k']} | aspect={inp['aspect_ratio']} (unique={pts_u.shape[0]})")
    ax.set_xlabel("x (world)")
    ax.set_ylabel("y (world)")
    plt.show()

# -----------------------------
# Batch runner
# -----------------------------
results = []

def run_batch_display(n=5, aspect_ratio=1.5, k=6, hotspot_count=10, seed0=0, temperature=0.2):
    global results
    results = []

    for i in range(n):
        inp = make_input(aspect_ratio=aspect_ratio, k=k, hotspot_count=hotspot_count, seed=seed0 + i)
        prompt = build_prompt(inp)

        completion_text = call_model_constrained(prompt, k=inp["k"], temperature=temperature)
        out_json = extract_output_json_constrained(completion_text)

        results.append({"case": i, "input": inp, "output": out_json})

        print("=" * 80)
        print(f"CASE {i}")
        print("INPUT_JSON:")
        print(json.dumps(inp, indent=2))
        print("OUTPUT_JSON:")
        print(json.dumps(out_json, indent=2))

        plot_voronoi(inp, out_json)

run_batch_display(n=8, aspect_ratio=1, k=10, hotspot_count=20, seed0=42)
```

```python
import json
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import Voronoi

def _wrap_json_preserve_indent(s: str, width: int = 92) -> list[str]:
    out = []
    for line in s.splitlines():
        if len(line) <= width:
            out.append(line)
            continue
        indent = len(line) - len(line.lstrip(" "))
        prefix = " " * indent
        chunk = line.strip()
        while len(chunk) > width - indent:
            out.append(prefix + chunk[: width - indent])
            chunk = chunk[width - indent :]
        if chunk:
            out.append(prefix + chunk)
    return out

def _case_text_lines(case, wrap_width: int):
    inp = case["input"]
    out = case["output"]
    lines = []
    lines += [f"CASE {case['case']}", "", "INPUT_JSON:"]
    lines += _wrap_json_preserve_indent(json.dumps(inp, indent=2), wrap_width)
    lines += ["", "OUTPUT_JSON:"]
    lines += _wrap_json_preserve_indent(json.dumps(out, indent=2), wrap_width)
    return lines

def _draw_voronoi(ax, inp, out, title_prefix=""):
    aspect = float(inp["aspect_ratio"])
    centers = out["centers"]
    hotspots = inp["hotspots"]

    pts = np.array([[c["x"] * aspect, c["y"]] for c in centers], dtype=float)
    xmin, xmax, ymin, ymax = 0.0, aspect, 0.0, 1.0

    vor = Voronoi(pts)
    regions, vertices = voronoi_finite_polygons_2d(vor, radius=max(aspect, 1.0) * 10.0)

    ax.set_aspect("equal", adjustable="box")
    ax.set_xlim(xmin, xmax)
    ax.set_ylim(ymin, ymax)

    for reg in regions:
        poly = vertices[reg]
        clipped = clip_polygon_to_bbox(poly, xmin, xmax, ymin, ymax)
        if clipped.shape[0] < 3:
            continue
        ax.fill(clipped[:, 0], clipped[:, 1], alpha=0.15, linewidth=0)
        ax.plot(
            np.r_[clipped[:, 0], clipped[0, 0]],
            np.r_[clipped[:, 1], clipped[0, 1]],
            linewidth=1,
        )

    ax.scatter(pts[:, 0], pts[:, 1], marker="x")

    for h in hotspots:
        hx, hy = h["x"] * aspect, h["y"]
        ax.scatter([hx], [hy], s=2000 * float(h["weight"]), alpha=0.5)
        ax.add_patch(plt.Circle((hx, hy), float(h["radius"]) * aspect, fill=False, linewidth=1))

    ax.set_title(f"{title_prefix}k={inp['k']} | aspect={inp['aspect_ratio']}", fontsize=9)
    ax.set_xlabel("x (world)")
    ax.set_ylabel("y (world)")

def save_single_page_montage_pdf(
    pdf_path="voronoi_montage_single_page.pdf",
    wrap_width=92,
    font_size=7.3,
    page_width_in=9,
    margin_in=0.25,
    col_gap_in=0.10,
    row_gap_in=0.00,         # set to 0 to remove whitespace between cases
    min_plot_height_in=2.0,
    text_col_frac=0.56,
):
    if not results:
        raise ValueError("results is empty. Run your batch generation first.")

    # Estimate per-line height in inches for monospace
    line_h_in = (font_size * 1.22) / 72.0

    items = []
    for case in results:
        lines = _case_text_lines(case, wrap_width)
        text_h = len(lines) * line_h_in + 0.16
        row_h = max(min_plot_height_in, text_h)
        items.append((case, lines, row_h))

    total_h = (2 * margin_in) + sum(h for _, _, h in items) + row_gap_in * max(0, len(items) - 1)

    fig = plt.figure(figsize=(page_width_in, total_h), dpi=200)
    fig.patch.set_facecolor("white")

    usable_w = page_width_in - 2 * margin_in
    text_w = usable_w * text_col_frac
    plot_w = usable_w - text_w - col_gap_in

    def ncoords(x, y, w, h):
        # Convert inches to normalized figure coords
        return [x / page_width_in, y / total_h, w / page_width_in, h / total_h]

    y_top = total_h - margin_in
    for case, lines, row_h in items:
        y0 = y_top - row_h

        ax_text = fig.add_axes(ncoords(margin_in, y0, text_w, row_h))
        ax_plot = fig.add_axes(ncoords(margin_in + text_w + col_gap_in, y0, plot_w, row_h))

        ax_text.axis("off")
        ax_text.text(
            0.0, 1.0,
            "\n".join(lines),
            va="top", ha="left",
            family="monospace",
            fontsize=font_size,
            transform=ax_text.transAxes,
            clip_on=False,
        )

        _draw_voronoi(ax_plot, case["input"], case["output"], title_prefix=f"Voronoi Case {case['case']} | ")

        y_top = y0 - row_gap_in

    fig.savefig(pdf_path)
    plt.close(fig)
    print(f"Saved single-page montage PDF: {pdf_path}")

def _fmt_num(x: float) -> str:
    # 1.5 -> 1.5, 1.0 -> 1, 0.333333 -> 0.3333
    return f"{float(x):g}"

def build_auto_filename_from_results(prefix="k", suffix=".pdf"):
    if not results:
        raise ValueError("results is empty.")

    inp0 = results[0]["input"]
    k = int(inp0["k"])
    a = float(inp0["aspect_ratio"])
    h = len(inp0.get("hotspots", []))

    return f"{prefix}{k}-h{h}-a{_fmt_num(a)}{suffix}"

# Use it:
auto_name = build_auto_filename_from_results(prefix="k", suffix=".pdf")
save_single_page_montage_pdf(pdf_path=auto_name)
print("Saved as:", auto_name)
```
